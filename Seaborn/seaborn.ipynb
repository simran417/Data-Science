```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
# to check the data types:
df.dtypes

# to find correlation:
correlation_matrix = df.select_dtypes(include=['number']).corr()
print(correlation_matrix)
```

```python
# heatmaps:
sns.heatmap(correlation_matrix)
```

```python
# join plot:
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')

# when kind is reg:
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

```python
# pair plot:
sns.pairplot(df)
```

```python
# to make graph using specified column we use hue:
sns.pairplot(df, hue='sex')
```

```python
# to count values
df['smoker'].value_counts()
```

```python
# dist plot:
sns.displot(df['tip'])

# if kde shoud be false:  kde=kernel density distribution
sns.displot(df['tip'],kde=False, bins=10)
```

```python
#countplot:
sns.countplot(x='sex', data=df)
```

```python
# with respect to day:
sns.countplot(x='day', data=df)
```

```python
# for vertical:
# with respect to day:
sns.countplot(y='day', data=df)
```

```python
# bar plot:
sns.barplot(x='total_bill', y='sex',data=df)
```

```python
# box plot:
sns.boxplot(x='smoker',y='total_bill',data=df)   #to give colors we use palette
```

```python
# box plot on complete dataset:  we also use hue in this
sns.boxplot(data=df,orient='v')
```

```python
# violin plot:
sns.violinplot(x='total_bill', y='day',data=df,palette='rainbow')
```
