```python
# merge operation:  (values must be same), (based on columns)
np.random.seed(23)
S1=np.random.randn(3,4)
np.random.seed(23)
S2=np.random.randn(3,4)
Df1=pd.DataFrame(S1,columns=['a','b','c','d'], index=[0,1,2])
Df2=pd.DataFrame(S2,columns=['a','b','c','d'], index=[3,4,5])
Df1
Df2
pd.merge(Df1,Df2)
pd.merge(Df1,Df2, on="a")  #we can take 2 or more input in on
# pd.merge(Df1,Df2, on="a",how=Df1)  #how means all the data of df1 and take only matched data of df2, how=outer means all the data of right and left table
```

```python
# join operation: (based on indexes)
Df1=pd.DataFrame(S1,columns=['a','b','c','d'], index=[0,1,2])
Df2=pd.DataFrame(S2,columns=['e','f','g','h'], index=[3,4,5])
Df1.join(Df2)
# Df1.join(Df2,how=Df1)
```
