```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[10,20,30,40]
y=["c","java","cpp","python"]
ex=[0.4,0.0,0.0,0.0]
c=["pink","purple","blue","green"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%", shadow=True,radius=1,labeldistance=1.2,startangle=90,textprops={"fontsize":15}
        ,wedgeprops={"linewidth":4,"edgecolor":"green"}, center=(3,6),) #rotatelabels=True   #by default startangle is 0
plt.legend()
plt.show()
```

```python
# dot piechart:
plt.pie([1])
plt.show()
```

```python
# to make multiple pie chart in one chart:
x1=[10,20,30,40]
x2=[2,4,6,8]
y=["c","java","cpp","python"]
plt.pie(x1,labels=y,radius=1.5)
plt.pie(x2,radius=0.5)
plt.show()
```

```python
# to make ring type chart:   also called as donat pie chart:
x1=[10,20,30,40]
x2=[2,4,6,8]
y=["c","java","cpp","python"]
plt.pie(x1,labels=y,radius=1.5)
plt.pie([1],colors="white")
plt.show()

# another method:
cr=plt.circle(xy=(0,0),radius=1,facecolor="w")
plt.gca.add_artist(cr)
plt.show()
```
