```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[2,4,6,8]
y=[2,5,4,6]
plt.stem(x,y , linefmt=":", markerfmt="+", bottom=1,basefmt="pink",label="python", use_line_collection=False, orientation="horizontal") #: for dotted line, if bottom =9 then the graph is reverse
plt.legend()
plt.show()
```

**Box Plot and whisker Plot**

```python
x=[10,20,20,40,50]
# plt.boxplot(x)
# plt.boxplot(x,notch=True)
# to change the orientation:
# plt.boxplot(x,vert=False)
plt.boxplot(x,labels=["python"],patch_artist=True,showmeans=True)
plt.show()
```

```python
# for multiple box plots:
x=[10,20,20,40,50]
y=[30,40,50,60,70]
z=[x,y]
plt.boxplot(z)
plt.show()
```
