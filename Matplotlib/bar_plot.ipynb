```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=['python','cpp','c','java']
y=[89,67,50,70]
plt.xlabel("languages",fontsize=15)  #to add labels
plt.ylabel("popularity",fontsize=15)
plt.title("popularity of languages")    #to add title
c=['yellow','green','grey','pink']
plt.bar(x,y,width=0.5,color=c,edgecolor='black',label="first")              # linewidth=3,linestyle=":",alpha=0.5
plt.legend()    #without this label is not work
plt.show()
```

```python
# to make multiple bargraphs in one graph:
x=['python','cpp','c','java']
y=[89,67,50,70]
z=[20,30,40,50]
plt.xlabel("languages",fontsize=15)  #to add labels
plt.ylabel("popularity",fontsize=15)
plt.title("popularity of languages")    #to add title
# c=['yellow','green','grey','pink']
plt.bar(x,y,width=0.5,color="pink",edgecolor='black',label="first")              # linewidth=3,linestyle=":",alpha=0.5
plt.bar(x,z,width=0.5,color="grey",edgecolor='black',label="second") 
plt.legend()    #without this label is not work
plt.show()
```

```python
# to seprates these graphs:
x=['python','cpp','c','java']
y=[89,67,50,70]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=(j+width for j in p)

plt.xlabel("languages",fontsize=15)  #to add labels
plt.ylabel("popularity",fontsize=15)
plt.title("popularity of languages")    #to add title

plt.bar(p,y,width,color="pink",label="first")           
plt.bar(p1,z,width,color="grey",label="second") 
plt.xticks(p+width/2,x,rotation=20)
plt.legend()    #without this label is not work
plt.show()
```

```python
# to make horizontal bar graph:
x=['python','cpp','c','java']
y=[89,67,50,70]
z=[20,30,40,50]
plt.xlabel("languages",fontsize=15)  #to add labels
plt.ylabel("popularity",fontsize=15)
plt.title("popularity of languages")    #to add title
# c=['yellow','green','grey','pink']
plt.barh(x,y,color="pink",label="first")              # to make horizontal graphs
plt.barh(x,z,color="grey",label="second") 
plt.legend()    #without this label is not work
plt.show()
```
