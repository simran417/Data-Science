```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4,5,6]
y=[1,2,3,4,5,6]
c=['red','blue','green','pink','orange','yellow']
s=[400,300,200,100,300,400]
plt.scatter(x,y,color=c,sizes=s,alpha=0.5)
plt.title("scatter plot",fontsize=15)
plt.xlabel("x-axis",fontsize=15)
plt.ylabel("y-axis",fontsize=15)
plt.show()
```
