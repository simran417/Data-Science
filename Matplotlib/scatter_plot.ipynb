```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4,5,6]
y=[1,2,3,4,5,6]
c=['red','blue','green','pink','orange','yellow']
s=[400,300,200,100,300,400]
plt.scatter(x,y,color=c,sizes=s,alpha=0.5,marker='*',edgecolors='black')   #marker is used to change the shape of scatter elements
plt.title("scatter plot",fontsize=15)
plt.xlabel("x-axis",fontsize=15)
plt.ylabel("y-axis",fontsize=15)
plt.show()
```

```python
# cmap function : used for color
x=[1,2,3,4,5,6]
y=[1,2,3,4,5,6]
z=[2,4,6,8,9,10]
c=[10,49,30,29,56,60]
s=[400,300,200,100,300,400]
plt.scatter(x,y,c=c,s=s,cmap="viridis")   
plt.scatter(x,z,c=c,s=s)
t=plt.colorbar()   #to show the color range
t.set_label("colorbar")
plt.title("scatter plot",fontsize=15)
plt.xlabel("x-axis",fontsize=15)
plt.ylabel("y-axis",fontsize=15)
plt.show()
```
